# Compare Apriori vs FP-Growth

# 5

```python
import pandas as pd
import matplotlib.pyplot as plt

ap = pd.read_csv("../data/processed/rules_apriori_filtered.csv")
fp = pd.read_csv("../data/processed/rules_fpgrowth_filtered.csv")

print("Apriori rules:", len(ap))
print("FP-Growth rules:", len(fp))
```

```python
plt.hist(ap["lift"], bins=30, alpha=0.6, label="Apriori")
plt.hist(fp["lift"], bins=30, alpha=0.6, label="FP-Growth")
plt.legend()
plt.title("Lift Distribution Comparison")
plt.show()
```
